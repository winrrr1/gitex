## 선형 독립과 선형 종속

```python
import numpy as np

# 벡터 집합 정의
# 예를 들어, 두 벡터 [3, 6]과 [6, 12]를 확인합니다.
vectors = np.array([[3, 6], 
                    [6, 12]])

# 벡터 집합의 랭크를 계산
rank = np.linalg.matrix_rank(vectors)

# 벡터 집합의 개수
num_v = vectors.shape[0]

# 선형 독립성 여부를 확인 위의 예제는 rank = 1 , num_v = 2로 다르기 때문에 선형종속 
if rank == num_v:
    print("선형 독립")
else:
    print("선형 종속")
```

```python
# 주어진 벡터를 행렬로 표현
A = np.array([
    [2, 7],
    [5, 1]
])
print(A)

# 행렬의 랭크를 계산
rank = np.linalg.matrix_rank(A)

# 벡터의 수
num_vectors = A.shape[1]

# 랭크와 벡터의 수를 비교하여 선형 독립 여부 확인
if rank == num_vectors:
    print("주어진 벡터는 선형 독립입니다.")
else:
    print("주어진 벡터는 선형 종속입니다.")
```
