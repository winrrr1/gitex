```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc

#   
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows 

#     
plt.rcParams['axes.unicode_minus'] = False

#  
v = np.array([3, 4])

#  
origin = np.array([0, 0])

#  
plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color='r', label=' [3, 4]')

#    
plt.plot([0, v[0]], [0, 0], 'bo--', label='x')
plt.plot([v[0], v[0]], [0, v[1]], 'bo--', label='y')

#   
plt.xlim(-1, 5)
plt.ylim(-1, 5)

#    
plt.title(' [3, 4]  ')
plt.xlabel('x')
plt.ylabel('y')

#   
plt.grid()
plt.legend()

#   
plt.text(1.5, 2, ' = 5')
```

```python
import numpy as np

#   
#  ,   [3, 6] [6, 12] .
vectors = np.array([[3, 6], 
                    [6, 12]])

#    
rank = np.linalg.matrix_rank(vectors)

#   
num_v = vectors.shape[0]

#       rank = 1 , num_v = 2    
if rank == num_v:
    print(" ")
else:
    print(" ")
```
