## 지수함수

```python
import math
```

```python
# 파이썬에서 제공하는 지수함수를 유형별로 실행
# 2의 5제곱
pow = math.pow(2,2)
print("pow 결과:",pow)
```

```python
# 2의 제곱근(루트값계산)
sqrt = math.sqrt(2)
print("sqrt 결과:",sqrt)
```

```python
# e의 2제곱
exp = math.exp(2)
print("exp의 결과:",exp)
```

```python
# a > 0 
a=2
math.pow(a,2)
```

```python
a**2
```

```python
# a < 0
a = -2
math.pow(a,-2)
```

```python
# 0 < a < 1
a = 0.5
math.pow(a,0.5)
```

```python
# 1번은 지수에 문자가 있어야 하므로 지수함수가 아닙니다. 
# 2번은 지수에 문자가 있고 밑이 0보다 크므로 지수함수 입니다. 2번만 지수함수 
# 3번은 밑은 마이너스 일수 없으므로 지수함수가 아닙니다. 
# 4번은 밑이 1이 아니어야 하므로 지수함수가 아닙니다.
```

```python
import numpy as np

# 1차원 배열
arr1 = np.arange(1, 9+1)
print("1차원 배열 (arr1):")
print(arr1)

# 2차원 배열
arr2 = np.arange(1, 9+1).reshape(3, 3)
print("\n2차원 배열 (arr2):")
print(arr2)

# 3차원 배열
arr3 = np.arange(1, 9+1).reshape(1, 3, 3)
print("\n3차원 배열 (arr3):")
print(arr3)
```
