```python
from sklearn.datasets import make_classification  # 분류용 데이터셋을 생성하기 위한 함수 임포트
from sklearn.model_selection import train_test_split  # 데이터셋을 학습용과 테스트용으로 나누기 위한 함수 임포트
from sklearn.linear_model import LogisticRegression  # 로지스틱 회귀 모델 임포트
from sklearn.metrics import ConfusionMatrixDisplay  # 혼동 행렬을 시각화하기 위한 클래스 임포트

# 분류 문제를 위한 데이터셋 생성. n_samples=10_000으로 10,000개의 샘플을 생성하고, 클래스 불균형을 위해 weights=[0.9, 0.1]을 설정
X, y = make_classification(n_samples=10_000, weights=[0.9, 0.1], random_state=0)

# 생성된 데이터셋을 학습용과 테스트용으로 나누기. random_state=0으로 랜덤 시드를 고정하여 결과의 재현성을 보장
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# 로지스틱 회귀 모델을 C=1e6으로 설정하여 학습. C 값은 정규화의 강도를 조절하며, 큰 값은 정규화 강도를 낮춤
classifier_05 = LogisticRegression(C=1e6, random_state=0).fit(X_train, y_train)

# 테스트 데이터셋을 사용하여 예측한 결과를 바탕으로 혼동 행렬을 시각화
# _ = ConfusionMatrixDisplay.from_estimator(classifier_05, X_test, y_test)
```
