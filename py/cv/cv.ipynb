```python
import cv2 as cv

# 흑백 이미지 읽기
img_gg = cv.imread('./img/cat_g.jpg', flags=cv.IMREAD_GRAYSCALE)

# 흑백 이미지를 컬러(BGR) 이미지로 변환
color_image = cv.cvtColor(img_gg, cv.COLOR_GRAY2BGR)

# 변환된 컬러 이미지 화면에 표시
cv.imshow('cat', color_image)

# 변환된 컬러 이미지를 파일로 저장
cv.imwrite('./img/cat_c.jpg', color_image)

# 5초 동안 대기 후 모든 윈도우 종료
cv.waitKey(5000)
cv.destroyAllWindows()

```

```python
import cv2 as cv

# 비디오 파일 열기
video = './img/big_buck.avi'  # 비디오 파일의 경로를 설정하세요.
cap = cv.VideoCapture(video)

# 비디오 파일이 제대로 열렸는지 확인
if not cap.isOpened():
    print("Error")
    exit()

while True:
    # 비디오에서 프레임 읽기
    ret, frame = cap.read()
    
    # 프레임을 제대로 읽지 못했으면 비디오 끝
    if not ret:
        break

    # 프레임 표시
    cv.imshow('vig buck', frame)

    # 'q' 키를 누르면 비디오를 중지
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

# 비디오 캡처 객체 해제 및 모든 윈도우 종료
cap.release()
cv.destroyAllWindows()
```
