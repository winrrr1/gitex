```python
import cv2 as cv

#   
video = './img/big_buck.avi'  #    .
cap = cv.VideoCapture(video)

#     
if not cap.isOpened():
    print("Error")
    exit()

while True:
    #   
    ret, frame = cap.read()
    
    #      
    if not ret:
        break

    #  
    cv.imshow('vig buck', frame)

    # 'q'    
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

#        
cap.release()
cv.destroyAllWindows()
```
