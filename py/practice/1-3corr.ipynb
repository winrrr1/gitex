## 첫번재 머신러닝 프로그램

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

```python
length = bream_length+smelt_length
weight = bream_weight+smelt_weight
```

```python
fish_data = [[l,w] for l,w in zip(length, weight)]
print(fish_data)
```

```python
len(fish_data)
```

```python
fish_target = [1]*35 + [0]*14
print(fish_target)
```

```python
len(fish_target)
```

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)
kn.score(fish_data, fish_target)
```

```python
# k최근접 이웃 알고리즘 
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30, 600, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 새로운 생선예측 
kn.predict([[30,600]]) # 돔 프레딕트 리스트 생선 크기 가로, 세로 
```

```python
# 새로운 생선예측 
kn.predict([[15,50]]) # 빙어
```

```python
kn49 = KNeighborsClassifier(n_neighbors=49) # 49로 설정할때 1, 100%가 아님 

kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```

```python
# 확인 문제
# 1번 특성 : 데이터를 표현하는 하나의 성질, 국가데이터의 경우 인구수, GDP,면적등이 하나의 국가를 나타냅니다 
# 2번 KNeighborsClassifier : 가장 가까운 이웃을 참고하여 정답을 예측하는 알고리즘이 구현된 사이킷런 클래스
# 3번 fit : 사이킷런 모델을 훈련할때 사용하는 메서드는 
# 4번 답

kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
    # 최근접 이웃 개수 설정
    kn.n_neighbors = n
    print(kn.n_neighbors)
    
    # 점수 계산
    score = kn.score(fish_data, fish_target)
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    if score < 1:
        print(n, score)
        break
```
