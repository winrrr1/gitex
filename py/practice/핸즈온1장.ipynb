## 연습문제(1)

### 1.머신러닝을 어떻게 정의할 수 있나?
##### 머신러닝(Machine Learning)은 컴퓨터 시스템이 명시적인 프로그래밍 없이 데이터를 통해 학습하고, 예측하거나 결정을 내릴 수 있도록 하는 기술입니다. 이는 알고리즘을 사용하여 데이터에서 패턴과 규칙을 자동으로 학습하고, 이를 바탕으로 새로운 데이터에 대해 예측하거나 결정을 내릴 수 있도록 합니다.

### 2.머신러닝이 도움을 줄 수 있는 문제 유형 네 가지
##### 분류(Classification): 데이터가 사전에 정의된 클래스 중 하나로 분류되는 문제 (예: 이메일 스팸 필터링).
##### 회귀(Regression): 연속적인 값 예측 문제 (예: 주택 가격 예측).
##### 군집화(Clustering): 비슷한 데이터 포인트를 그룹으로 묶는 문제 (예: 고객 세분화).
##### 강화 학습(Reinforcement Learning): 에이전트가 환경과 상호작용하면서 보상을 최대화하도록 학습하는 문제 (예: 자율주행차).

### 3.레이블된 훈련 세트란 무엇인가?
##### 레이블된 훈련 세트는 각 데이터 샘플에 대해 정답(레이블)이 포함된 데이터셋입니다. 지도 학습에서는 이러한 레이블을 사용하여 입력과 출력 간의 관계를 학습합니다. 예를 들어, 이미지 분류에서는 각 이미지에 해당하는 클래스 레이블이 포함된 데이터셋을 의미합니다.

### 4.가장 널리 사용되는 지도 학습 작업 두 가지는?
##### 분류(Classification): 입력 데이터를 미리 정의된 카테고리 중 하나로 분류하는 작업 (예: 스팸 이메일 분류).
##### 회귀(Regression): 연속적인 값 예측 작업 (예: 주식 가격 예측).

### 5.보편적인 비지도 학습 작업 네 가지는?
##### 군집화(Clustering): 데이터를 유사한 특성에 따라 그룹으로 묶는 작업 (예: 고객 세분화).
##### 차원 축소(Dimensionality Reduction): 데이터의 차원을 줄여 데이터의 주요 특징을 유지하면서 단순화하는 작업 (예: PCA).
##### 연관 규칙 학습(Association Rule Learning): 데이터 내의 항목 간의 관계를 발견하는 작업 (예: 장바구니 분석).
##### 밀도 추정(Density Estimation): 데이터의 분포를 모델링하여 새로운 데이터 샘플의 가능성을 평가하는 작업 (예: 히스토그램 기반 밀도 추정).

### 6.사전 정보가 없는 여러 지형에서 로봇을 걸어가게 하려면 어떤 종류의 머신러닝 알고리즘을 사용할 수 있나?
##### 강화 학습(Reinforcement Learning): 로봇이 환경과 상호작용하면서 최적의 경로를 학습할 수 있게 해주는 알고리즘입니다. 로봇은 보상 신호를 통해 학습하며, 탐색과 활용의 균형을 맞추면서 다양한 지형에서의 최적 행동을 학습합니다.

### 7.고객을 여러 그룹으로 분할하려면 어떤 알고리즘을 사용해야 하나?
##### 군집화(Clustering) 알고리즘: 고객 데이터를 유사한 특성에 따라 그룹으로 나누는 데 사용됩니다. 대표적인 군집화 알고리즘으로는 K-평균(K-Means), 계층적 군집화(Hierarchical Clustering), DBSCAN 등이 있습니다.

### 8.스팸 감지의 문제는 지도 학습과 비지도 학습 중 어떤 문제로 볼 수 있나?
##### 지도 학습(Supervised Learning): 스팸 감지 문제는 레이블된 데이터(스팸, 비스팸)로 학습하는 지도 학습 문제입니다. 모델은 입력된 이메일이 스팸인지 아닌지를 예측하기 위해 훈련됩니다.

### 9.온라인 학습 시스템이 무엇인가?
##### 온라인 학습(Online Learning): 데이터가 순차적으로 도착하는 상황에서 학습을 진행하는 시스템입니다. 모델은 새로운 데이터를 받으면서 계속해서 업데이트되며, 대규모 데이터셋을 처리하거나 지속적으로 변화하는 데이터에 적응하는 데 유용합니다.

### 10.외부 메모리 학습이 무엇인가?
##### 외부 메모리 학습(External Memory Learning): 모델이 주어진 메모리(외부 저장소)를 사용하여 데이터를 저장하고, 필요한 때에 해당 데이터를 참조하거나 업데이트하는 학습 방식입니다. 이는 대용량 데이터셋을 효과적으로 처리하고, 메모리 사용을 최적화하는 데 도움을 줍니다.

### 11.예측을 하기 위해 유사도 측정에 의존하는 학습 알고리즘은 무엇인가?
##### K-최근접 이웃(K-Nearest Neighbors, KNN) 알고리즘은 예측을 하기 위해 데이터 샘플 간의 유사도를 측정합니다. 새로운 데이터 포인트의 레이블은 가장 가까운 K개의 이웃 데이터 포인트의 레이블에 기반하여 결정됩니다.

### 12.모델 파라미터와 학습 알고리즘의 하이퍼파라미터 사이에는 어떤 차이가 있나?
##### 모델 파라미터: 학습 과정에서 모델이 자동으로 학습하는 값들입니다. 예를 들어, 선형 회귀에서는 기울기와 절편이 모델 파라미터입니다.
##### 하이퍼파라미터: 학습 과정 전에 사용자가 설정하는 값들입니다. 예를 들어, K-최근접 이웃 알고리즘의 K 값, 신경망의 층 수 및 노드 수, 학습률 등이 하이퍼파라미터입니다. 하이퍼파라미터는 모델의 학습 및 성능에 영향을 미치며, 보통 교차 검증 등을 통해 최적의 값을 찾습니다.

### 13.모델 기반 알고리즘이 찾는 것은 무엇인가요? 성공을 위해 이 알고리즘이 사용하는 가장 일반적인 전략은 무엇인가?  예측은 어떻게 만드나?
##### 모델 기반 알고리즘은 주어진 데이터를 기반으로 예측 또는 결정을 내리기 위해 데이터의 패턴과 구조를 모델링합니다. 이러한 알고리즘은 일반적으로 최적화(Optimization) 전략을 사용하여 모델 파라미터를 조정합니다. 가장 일반적인 전략은 **경량화된 최적화 기법(Gradient Descent)**과 **정규화(Regularization)**입니다. 이들은 데이터에 대해 최적의 모델을 학습하고 과적합을 방지하기 위해 사용됩니다.

##### 예측을 만드는 과정은 모델이 학습한 패턴을 사용하여 새로운 입력 데이터에 대한 출력을 생성하는 것입니다. 일반적인 과정은 다음과 같습니다:
###### 입력 데이터 준비: 모델에 예측할 데이터를 입력합니다.
###### 특징 추출: 입력 데이터에서 중요한 특징을 추출합니다.
###### 모델 적용: 학습된 모델을 사용하여 입력 데이터에 대한 예측을 수행합니다.
###### 출력 생성: 모델의 출력을 기반으로 예측 결과를 생성합니다.


### 14.머신러닝의 주요 도전 과제는 무엇인가?
##### 머신러닝의 주요 도전 과제
##### 데이터 품질과 양:데이터 부족, 불균형, 노이즈 문제로 모델 성능이 저하될 수 있습니다.
##### 모델의 일반화:모델이 훈련 데이터에 과적합하거나 과소적합하여 새로운 데이터에서 잘 작동하지 않을 수 있습니다.
##### 특징 선택 및 엔지니어링:중요한 특징을 선택하고 유용한 특징을 생성하는 것이 어려울 수 있습니다.
##### 계산 자원과 시간:대규모 데이터와 복잡한 모델을 학습하는 데 필요한 계산 자원과 시간이 많이 소요될 수 있습니다.
##### 모델의 해석 가능성:모델이 블랙박스처럼 작동하여 결과를 이해하거나 설명하기 어려울 수 있습니다.
##### 데이터 프라이버시와 윤리:개인 정보 보호와 윤리적 문제를 해결하는 것이 중요합니다.
##### 하이퍼파라미터 조정:최적의 하이퍼파라미터를 찾는 것이 어렵고 모델 성능에 큰 영향을 미칠 수 있습니다.

### 15.모델이 훈련 데이터에서의 성능은 좋지만 새로운 샘플에서의 일반화 성능이 나쁘다면 어떤 문제가 있나? 가능한 해결책 세 가지는 무엇인가?
##### 문제: 모델이 훈련 데이터에 과적합(overfitting)되어 새로운 데이터에서의 일반화 성능이 저하된 경우입니다.
##### 해결책: 
###### 정규화(Regularization): L1, L2 정규화와 같은 기법을 사용하여 모델의 복잡성을 줄이고 과적합을 방지합니다.
###### 더 많은 데이터 수집: 더 많은 훈련 데이터를 수집하여 모델이 다양한 데이터 패턴을 학습할 수 있게 합니다.
###### 교차 검증(Cross-validation): 데이터를 여러 개의 서브셋으로 나누어 모델을 평가하고 하이퍼파라미터를 조정하여 과적합을 방지합니다.

### 16.테스트 세트가 무엇이고 왜 사용해야 하나?
##### 테스트 세트는 모델이 학습되지 않은 새로운 데이터로, 모델의 최종 성능을 평가하는 데 사용됩니다. 테스트 세트는 훈련 과정과 검증 과정에서 사용되지 않으며, 모델이 실제 환경에서 얼마나 잘 작동하는지를 평가하는 데 도움을 줍니다.

### 17.검증 세트의 목적은 무엇인가?
##### 검증 세트는 모델을 훈련할 때 하이퍼파라미터를 조정하고 모델 성능을 평가하는 데 사용됩니다. 검증 세트는 훈련 데이터와는 별도로 사용되며, 모델의 성능을 실험하고 조정하기 위해 사용됩니다.

### 18.훈련-개발 세트가 무엇인가? 언제 필요하고 어떻게 사용해야 하나?
##### 훈련-개발 세트는 데이터를 훈련 세트와 검증 세트로 나누는 과정에서, 모델의 성능을 검증하기 위해 사용되는 데이터입니다. 일반적으로 데이터셋을 훈련 세트, 검증 세트, 테스트 세트로 나누는 경우가 많으며, 훈련-개발 세트는 검증 세트의 역할을 하며, 하이퍼파라미터 조정 및 모델 개선에 사용됩니다.

### 19.테스트 세트를 사용해 하이퍼파라미터를 튜닝하면 어떤 문제가 생기나?
##### 테스트 세트를 하이퍼파라미터 튜닝에 사용하면 모델이 테스트 세트에 대해 과적합(overfit)될 수 있습니다. 이는 테스트 세트가 모델 선택 과정에 영향을 미치게 되어, 실제 성능 평가가 정확하지 않을 수 있습니다. 따라서 테스트 세트는 최종 성능 평가에만 사용하고, 하이퍼파라미터 튜닝 및 모델 선택은 검증 세트를 사용하여 진행하는 것이 좋습니다.
