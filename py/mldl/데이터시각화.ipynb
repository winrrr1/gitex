```python
x = np.linspace(-4,4,100)
y1 = x**3
y2 = 10*x**2-1

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
x = np.linspace(-4,4,100)
y1 = x**3
y2 = 10*x**2-1

plt.plot(x,y1,'o--r',x,y2,'*--b')
plt.show()
```

```python
# numpy  -4 4 100     .
x = np.linspace(-4, 4, 100)

# y1 x  .
y1 = x**3

# y2 10x^2 - 1 .
y2 = 10 * x**2 - 1

# y1   (o-r), y2   (*-b) .
plt.plot(x, y1, 'o-r', x, y2, '*-b')

# x  -1 1  (    ).
plt.xlim(-1, 1)

# x  -3 3  .
plt.xlim(-3, 3)

# x  'X' .
plt.xlabel('X')

# y  'Y' .
plt.ylabel('Y')

#   '' .
plt.title('')

#    'upper right' .
plt.legend(['$ y1= x^3$ ', '$ y2 =10x^2-1 $ '], loc='upper right')

#      .
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Figure  .
fig = plt.figure()

# 3D    .
ax = fig.add_subplot(111, projection='3d')

# x, y  .
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# x y   .
X, Y = np.meshgrid(x, y)

# Z   (: Z = sin(sqrt(X^2 + Y^2))).
Z = np.sin(np.sqrt(X**2 + Y**2))

# 3D  .
ax.plot_surface(X, Y, Z, cmap='viridis')

#   .
plt.show()
```
