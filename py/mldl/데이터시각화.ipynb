```python
import numpy as np
import pandas as pd
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5,5,0.1)
```

```python
y1 = x**2 -2
```

```python
y2 = 20*np.cos(x)**2
```

```python
# 첫 번째 그래프의 크기를 5x5로 설정하고, 첫 번째 그래프를 그립니다.
plt.figure(1, figsize=(5, 5))
plt.plot(x, y1)

# 두 번째 그래프의 크기를 10x5로 설정하고, 두 번째 그래프를 그립니다.
plt.figure(2, figsize=(10, 5))
plt.plot(x, y2)

# 모든 그래프를 화면에 표시합니다.
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x, y1)

plt.subplot(2,1,2)
plt.plot(x, y2)

plt.show()
```

```python
# 여러가지 기능 
# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 기호가 깨지지 않도록 설정합니다.
plt.rcParams['axes.unicode_minus'] = False

# 2행 1열의 첫 번째 서브플롯을 설정합니다.
plt.subplot(2, 1, 1)

# 첫 번째 서브플롯에 x와 y1의 데이터를 플로팅합니다.
plt.plot(x, y1)

# 첫 번째 서브플롯에 제목을 추가합니다.
plt.title('첫 번째 그래프')

# 첫 번째 서브플롯에 x축 라벨을 추가합니다.
plt.xlabel('X축')

# 첫 번째 서브플롯에 y축 라벨을 추가합니다.
plt.ylabel('Y1 값')

# 첫 번째 서브플롯에 그리드 추가합니다.
plt.grid(True)


# 2행 1열의 두 번째 서브플롯을 설정합니다.
plt.subplot(2, 1, 2)

# 두 번째 서브플롯에 x와 y2의 데이터를 플로팅합니다.
plt.plot(x, y2)

# 두 번째 서브플롯에 제목을 추가합니다.
plt.title('두 번째 그래프')

# 두 번째 서브플롯에 x축 라벨을 추가합니다.
plt.xlabel('X축')

# 두 번째 서브플롯에 y축 라벨을 추가합니다.
plt.ylabel('Y2 값')

# 두 번째 서브플롯에 그리드 추가합니다.
plt.grid(True)

# 전체 레이아웃을 자동 조정하여 서브플롯 간의 간격을 최적화합니다.
plt.tight_layout()

# 모든 서브플롯을 화면에 표시합니다.
plt.show()
```

```python
x = np.linspace(-4,4,100)
y1 = x**3
y2 = 10*x**2-1

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
x = np.linspace(-4,4,100)
y1 = x**3
y2 = 10*x**2-1

plt.plot(x,y1,'o--r',x,y2,'*--b')
plt.show()
```

```python
# numpy를 사용하여 -4부터 4까지 100개의 균일한 값으로 이루어진 배열을 생성합니다.
x = np.linspace(-4, 4, 100)

# y1은 x의 세제곱 값입니다.
y1 = x**3

# y2는 10x^2 - 1의 값입니다.
y2 = 10 * x**2 - 1

# y1을 빨간색 원형 마커(o-r)로, y2를 파란색 별표 마커(*-b)로 플롯합니다.
plt.plot(x, y1, 'o-r', x, y2, '*-b')

# x축의 범위를 -1부터 1까지로 설정합니다 (이후 범위 설정에 의해 덮어쓰여집니다).
plt.xlim(-1, 1)

# x축의 범위를 -3부터 3까지로 다시 설정합니다.
plt.xlim(-3, 3)

# x축 라벨을 'X'로 설정합니다.
plt.xlabel('X')

# y축 라벨을 'Y'로 설정합니다.
plt.ylabel('Y')

# 그래프의 제목을 '플롯'으로 설정합니다.
plt.title('플롯')

# 범례를 설정하고 위치를 'upper right'로 지정합니다.
plt.legend(['$ y1= x^3$ ', '$ y2 =10x^2-1 $ '], loc='upper right')

# 모든 설정을 반영하여 그래프를 화면에 표시합니다.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Figure 객체를 생성합니다.
fig = plt.figure()

# 3D 플롯을 위한 서브플롯을 생성합니다.
ax = fig.add_subplot(111, projection='3d')

# x, y 데이터를 생성합니다.
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# x와 y의 그리드 데이터를 생성합니다.
X, Y = np.meshgrid(x, y)

# Z 데이터를 계산합니다 (예: Z = sin(sqrt(X^2 + Y^2))).
Z = np.sin(np.sqrt(X**2 + Y**2))

# 3D 그래프를 그립니다.
ax.plot_surface(X, Y, Z, cmap='viridis')

# 그래프를 화면에 표시합니다.
plt.show()
```
