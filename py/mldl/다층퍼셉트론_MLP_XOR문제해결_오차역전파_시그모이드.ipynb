```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows의 경우

# 마이너스 기호가 깨지지 않도록 설정합니다.
plt.rcParams['axes.unicode_minus'] = False

# 가중치와 바이어스
w11 = np.array([-2, -2])
w12 = np.array([2, 2])
w2 = np.array([1, 1])
b1 = 3
b2 = -1
b3 = -1
```

```python
# 퍼셉트론
def MLP(x, w, b):
    y = np.sum(w * x) + b
    if y <= 0:
        return 0
    else:
        return 1
```

```python
# NAND 게이트
def NAND(x1, x2):
    return MLP(np.array([x1, x2]), w11, b1)

# OR 게이트


def OR(x1, x2):
    return MLP(np.array([x1, x2]), w12, b2)

# AND 게이트


def AND(x1, x2):
    return MLP(np.array([x1, x2]), w2, b3)

# XOR 게이트


def XOR(x1, x2):
    return AND(NAND(x1, x2), OR(x1, x2))
```

```python
# x1, x2 값을 번갈아 대입해 가며 최종 값 출력
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = XOR(x[0], x[1])
    print("입력 값: " + str(x) + " 출력 값: " + str(y))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows의 경우

# 마이너스 기호가 깨지지 않도록 설정합니다.
plt.rcParams['axes.unicode_minus'] = False

# 활성화 함수 정의

def sigmoid(x):  # 시그모이드 함수 : 입력값을 0과 1 사이로 변환하는 비선형 함수
    return 1 / (1 + np.exp(-x))

def elu(x, alpha=1.0):  # ELU 함수 : 양수는 그대로, 음수는 지수적으로 변환하여 출력하는 함수
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def tanh(x):  # 하이퍼볼릭 탄젠트 함수 : 입력값을 -1과 1 사이로 변환하는 비선형 함수
    return np.tanh(x)

def softplus(x):  # 소프트플러스 함수 : 모든 값에서 미분 가능한 ReLU의 부드러운 버전
    return np.log(1 + np.exp(x))

# x 값 범위 생성 (-5에서 5까지 400개의 점)
x = np.linspace(-5, 5, 400)

# 각 활성화 함수에 대한 y 값 계산
y_sigmoid = sigmoid(x)  # 시그모이드 함수에 대한 y 값
y_elu = elu(x)  # ELU 함수에 대한 y 값
y_tanh = tanh(x)  # 하이퍼볼릭 탄젠트 함수에 대한 y 값
y_softplus = softplus(x)  # 소프트플러스 함수에 대한 y 값

# 그래프 생성
plt.figure(figsize=(12, 8))  # 그림의 크기 설정 (12x8 인치)

# Sigmoid 함수 그래프
plt.subplot(2, 2, 1)  # 2x2 행렬의 첫 번째 위치에 그래프 배치
plt.plot(x, y_sigmoid, label='Sigmoid', color='blue')  # 파란색 시그모이드 그래프
plt.title('Sigmoid 함수')  # 그래프 제목
plt.grid(True)  # 격자 표시

# ELU 함수 그래프
plt.subplot(2, 2, 2)  # 2x2 행렬의 두 번째 위치에 그래프 배치
plt.plot(x, y_elu, label='ELU', color='green')  # 초록색 ELU 그래프
plt.title('ELU 함수')  # 그래프 제목
plt.grid(True)  # 격자 표시

# Tanh 함수 그래프
plt.subplot(2, 2, 3)  # 2x2 행렬의 세 번째 위치에 그래프 배치
plt.plot(x, y_tanh, label='Tanh', color='red')  # 빨간색 Tanh 그래프
plt.title('Tanh 함수')  # 그래프 제목
plt.grid(True)  # 격자 표시

# Softplus 함수 그래프
plt.subplot(2, 2, 4)  # 2x2 행렬의 네 번째 위치에 그래프 배치
plt.plot(x, y_softplus, label='Softplus', color='purple')  # 보라색 Softplus 그래프
plt.title('Softplus 함수')  # 그래프 제목
plt.grid(True)  # 격자 표시

plt.tight_layout()  # 그래프 간의 간격을 자동으로 조정
plt.show()  # 그래프 화면에 표시
```
