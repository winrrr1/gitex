```python
# 0에서 10까지 100개의 균등하게 분포된 값을 가진 배열 x를 생성합니다.
x = np.linspace(0, 10, 100)

# x에 대한 사인 함수를 계산하여 y에 저장합니다.
y = np.sin(x)

# x에 대한 코사인 함수를 계산하여 z에 저장합니다.
z = np.cos(x)

# 새로운 Figure 객체를 생성합니다.
fig0 = plt.figure()

# Figure에 새로운 Axes 객체를 추가합니다. 위치 및 크기는 [왼쪽, 아래, 너비, 높이]로 정의됩니다.
ax0 = fig0.add_axes([0, 0, 1, 1])

# 생성된 Axes에 x와 y 값을 사용하여 선 그래프를 플로팅합니다.
ax0.plot(x, y)

# 생성된 모든 그래프를 화면에 출력합니다.
plt.show()
```
