```python
import itertools
from collections import Counter

# 윷의 결과 상태
res = ['도', '개', '걸', '윷', '모']

# 윷 4개의 가능한 모든 조합 생성
combinations = list(itertools.product(res, repeat=4))

# 각 조합의 출현 빈도 수를 카운트
counter = Counter(combinations)

# 총 가능한 조합 수
total_combinations = len(combinations)

# 확률 계산 및 출력
probabilities = {comb: count / total_combinations for comb, count in counter.items()}

# 확률 출력
for comb, prob in probabilities.items():
    print(f"{comb}: {prob:.4f}")
```
