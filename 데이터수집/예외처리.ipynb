## 주식정보 가져오기

```python
import urllib.request
import re
```

```python
url = 'https://finance.naver.com/item/main.nhn?code=005930'
html = urllib.request.urlopen(url)
```

```python
html.read()
```

```python
html_content = str(html.read().decode("ms949"))
```

```python
html_content[0:1000]
```

```python
stock_results = re.findall(
    "(\<dl class=\"blind\"\>)([\s\S]+?)(\<\/dl\>)", html_content)
```

```python
samsung_stock = stock_results[0]
```

```python
samsung_stock
```

```python
samsung_index = samsung_stock[1]
```

```python
index_list = re.findall("(\<dd\>)([\s\S]+?)(\<\/dd\>)", samsung_index)
```

```python
for index in index_list:
    print(index[1])
```

```python
with open('naver_주식.txt', 'w') as nf:
    for index in index_list:
        nf.write(f"{index[1]}\n")  # index[1]을 파일에 작성하고 줄 바꿈 추가
```

## 비트코인 

```python
import requests
from bs4 import BeautifulSoup
```

```python
# 웹 페이지 요청 및 HTML 콘텐츠 가져오기
bcoin = requests.get('https://m.stock.naver.com/crypto/UPBIT/BTC')
```

```python
# BeautifulSoup을 사용하여 HTML 파싱
bss= BeautifulSoup(bcoin.content, 'html.parser')
```

```python
# 원하는 HTML 요소 찾기
lm = bss.find_all('li',{'class':'DetailInfoStatus_item__hFb9W'})
lm1 = bss.find_all('div',{'class':'DetailInfo_article__MxD3U'})
```

```python
print(lm)
print(lm1)
```

```python
# 텍스트를 저장할 빈 리스트 생성
lmm = []

# 각 요소에서 텍스트만 추출하여 리스트에 추가하고 출력
for i in lm:
    txt = i.get_text(strip=True)
    lmm.append(txt)  # 텍스트를 리스트에 추가
for j in lm1:
    txt1 = j.get_text(strip=True)
    lmm.append(txt1)  # 텍스트를 리스트에 추가
# lmm 리스트 확인
print(lmm)
```

```python
with open('bcoin.txt', 'w') as bcoin:
    bcoin.write('\n'.join(lmm))
```
